```python
a_list = [a_01, a_02, a_03, a_04, a_05, a_06, a_07, a_08, a_09, a_10, a_11, a_12, a_13, a_14, a_15,
           a_16, a_17, a_18, a_19, a_20, a_21, a_22, a_23, a_24, a_25, a_26, a_27, a_28, a_29,
           a_30, a_31, a_32]

# DataFrame 합치기
a = pd.concat(a_list, ignore_index=True)

a
```

```python
a = a.drop(columns = ["X", "distance", "Use_SoC", "max..SoC....",
                     "min..SoC....", "displayed.SoC....", "SoC....",
                     "Velocity..km.h.", "Time..s."])
```

```python
a = a.fillna(0)
```

```python
a = a.values  # DataFrame을 NumPy 배열로 변환
x = a[:, 0:22]  # 첫 번째 열부터 21번째 열까지 선택
y = a[:, 22]  # 22번째 열 선택
```

```python
from sklearn.model_selection import train_test_split

seed = 42
test_size = 0.3
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state = seed)
x_train
x_test
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=1983, max_depth=11, min_child_weight=25, gamma=1, colsample_bytree=0.6, reg_lambda=0.7560045700246312, reg_alpha=0.0010564184128526689, subsample=1.0)
xgb_model = model.fit(x_train, y_train, verbose=False, eval_set=[(x_test, y_test)])
```

```python
y_pred = model.predict(x_test)
predictions = [value for value in y_pred]
```

```python
import optuna

def objectiveXGB(trial):
    param = {
        "n_estimators" : trial.suggest_int('n_estimators', 500, 2000),  # 범위를 줄임
        'max_depth':trial.suggest_int('max_depth', 8, 12),  # 범위를 줄임
        'min_child_weight':trial.suggest_int('min_child_weight', 1, 100),  # 범위를 줄임
        'gamma':trial.suggest_int('gamma', 1, 3),
        'learning_rate': 0.01,
        'colsample_bytree':trial.suggest_float('colsample_bytree',0.5, 1, step=0.1),
        'nthread' : -1,
#       'tree_method': 'gpu_hist',
#        'predictor': 'gpu_predictor',
        'lambda': trial.suggest_float('lambda', 1e-3, 10.0, log=True),
        'alpha': trial.suggest_float('alpha', 1e-3, 10.0, log=True),
        'subsample': trial.suggest_categorical('subsample', [0.6,0.7,0.8,1.0] ),
        'random_state': 42
    }

    model = XGBRegressor(**param)
    model.fit(x_train, y_train, early_stopping_rounds=10, eval_set=[(x_test, y_test)])  # 조기 중단 사용
    y_pred = model.predict(x_test)
    error = mean_squared_error(y_test, y_pred)

    return error

study = optuna.create_study(direction="minimize")
study.optimize(objectiveXGB, n_trials=20)  # 시도 횟수를 줄임

print(study.best_params)
```
